# Popularity Based Recommender System


```python
rating_with_name = ratings.merge(books,on="ISBN")
rating_with_name 
```

```python
num_rating_df = rating_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns = {'Book-Rating':'Num-Rating'},inplace= True)
num_rating_df
```

```python
avg_rating_df = rating_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns = {'Book-Rating':'avg-Rating'},inplace = True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on="Book-Title")
popular_df
```

```python
popular_df = popular_df[popular_df['Num-Rating']>=250].sort_values('avg-Rating',ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','Num-Rating','avg-Rating']]
```

# Collaborative Filtering Based Recommender System

```python
x=rating_with_name.groupby('User-ID').count()['Book-Rating'] > 200
important_users = x[x].index
```

```python
filtered_rating = rating_with_name[rating_with_name['User-ID'].isin(important_users)]
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_book = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_book)]
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('You Belong To Me')
```
